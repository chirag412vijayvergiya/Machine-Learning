```python
x_train = np.array([1.0, 2.0])           #(size in 1000 square feet)
y_train = np.array([300.0, 500.0])   
```

```python
def compute_cost(x, y, w, b): 
    m = x.shape[0]
    cont_sum = 0
    for i in range(m):
        f_wb = w * x[i] + b
        error = f_wb - y[i]
        cost = (error)**2
        cont_sum += cost
    return cont_sum / (2 * m)

```

```python
def custom_plot(x, y):
    plt.figure()
    plt.scatter(x, y, marker='x', c='r', label='Training Data')
    plt.plot(x, y, label='Linear Fit')
    plt.xlabel('Size (1000 sqft)')
    plt.ylabel('Price (in 1000s of dollars)')
    plt.title('Housing Prices')
    plt.legend()
    plt.grid(True)
    plt.show()

custom_plot(x_train, y_train)
```
